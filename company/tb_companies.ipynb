```python
import os
import json
import pandas as pd
from urllib import request as req
```

```python
# set local data
STORAGE = os.path.join('..', 'storage')
FILENAME = 'tb_estabelecimento.1.1.0.csv'
```

```python
# read data from local
usecols = ['CNPJ', 'NOME_FANTASIA', 'CEP', 'TIPO_LOGRADOURO', 'LOGRADOURO', 'COMPLEMENTO', 'BAIRRO']
df = pd.read_csv(f'{STORAGE}/{FILENAME}', sep=';', dtype=str, usecols=usecols)
```

```python
def post_all(url: str, data) -> dict:
    api = req.Request(url, method="POST")
    api.add_header('Content-type', 'application/json')

    response = req.urlopen(api, data=bytes(data, encoding='utf-8'))
    raw_data = response.read()
    encoding = response.info().get_content_charset('utf8')

    return raw_data.decode(encoding)
```

```python
post_all('http://localhost:8080/massive', df.to_json(orient='records'))
```
